```python
uganda_p_df = uganda_p_df.set_index(['t','m'])
uganda_p_df.columns.name = 'j'

uganda_p_df = uganda_p_df.apply(lambda x: pd.to_numeric(x,errors='coerce'))
uganda_p_df = uganda_p_df.replace(0,np.nan)

uganda_fct_df

uganda_fct_df = uganda_fct_df.set_index('j')
uganda_fct_df.columns.name = 'n'

uganda_fct_df = uganda_fct_df.apply(lambda x: pd.to_numeric(x,errors='coerce'))


rdi_df = rdi_df.set_index('n')
rdi_df.columns.name = 'k'

pbar_uga = uganda_p_df.xs('2019-20',level='t').mean()
pbar_uga = pbar_uga[result_uganda.beta.index]

xhat_uga = result_uganda.predicted_expenditures()

# Total food expenditures per household
xbar_uga = xhat_uga.groupby(['i','t','m']).sum()

# Reference budget
xref_uga = xbar_uga.quantile(0.5)

def my_prices(p0,uganda_p_df=pbar_uga,j='Millet'):
    """
    Change price of jth good to p0, holding other prices fixed.
    """
    uganda_p_df = uganda_p_df.copy()
    uganda_p_df.loc[j] = p0
    return uganda_p_df

qhat_uga = (xhat_uga.unstack('j')/pbar_uga).dropna(how='all')

# Drop missing columns
qhat_uga = qhat_uga.loc[:,qhat_uga.count()>0]

def nutrient_demand(uganda_expenditure_df,uganda_p_df):
    c = result_uganda.demands(uganda_expenditure_df,uganda_p_df)
    fct0,c0 = uganda_fct_df.align(c,axis=0,join='inner')
    N = fct0.T@c0

    N = N.loc[~N.index.duplicated()]
    
    return N
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

my_j = 'Millet'  # Interesting Ugandan staple

P = np.geomspace(.01,10,50)*pbar_uga[my_j]

# Utility of median household, given prices
U0 = result_uganda.indirect_utility(xref_uga,pbar_uga)

plt.plot([result_uganda.demands(xref_uga,my_prices(p0,j=my_j))[my_j] for p0 in P],P)
plt.plot([result_uganda.demands(U0,my_prices(p0,j=my_j),type="Hicksian")[my_j] for p0 in P],P)
plt.ylabel('Price')
plt.xlabel(my_j)
plt.legend(("Marshallian","Hicksian"))
plt.title(f"Uganda's Marshallian and Hicksian Demand Curves for {my_j}")
```

```python
def compensating_variation(U0,p0,p1):
    x0 = result_uganda.expenditure(U0,p0)
    x1 = result_uganda.expenditure(U0,p1)

    return x1-x0

def revenue(U0,p0,p1,type='Marshallian'):
    """(Un)Compensated revenue from taxes changing vector of prices from p0 to p1.

    Note that this is only for *demand* side (i.e., if supply perfectly elastic).
    """
    
    dp = p1 - p0 # Change in prices

    c = result_uganda.demands(U0,p1,type=type)

    dp,c = dp.align(c,join='inner')

    return dp.T@c


def deadweight_loss(U0,p0,p1):
    """
    Deadweight loss of tax/subsidy scheme creating wedge in prices from p0 to p1.

    Note that this is only for *demand* side (i.e., if supply perfectly elastic).
    """
    cv = compensating_variation(U0,p0,p1)

    return cv - revenue(U0,p0,p1,type='Hicksian')
```

#### Price Changes, Revenue, and Compensating Variation
If price of a good increases/decreases, what’s the cost to the household?

The compensating variation associated with the price change is understood as the income change the household would need to make it just as well off as they were with the old prices.
This is plotted against revenue, which is what the tax or subsidy from changing the price would generate.

```python
fig, ax1 = plt.subplots()

ax1.plot(P,[compensating_variation(U0,pbar_uga,my_prices(p0,j=my_j)) for p0 in P])
ax1.set_xlabel(f"Price of {my_j}")
ax1.set_ylabel("Compensating Variation")

ax1.plot(P,[revenue(U0,pbar_uga,my_prices(p0,j=my_j),type='Hicksian') for p0 in P],'k')
ax1.legend(('Compensating Variation','Revenue'))
ax1.axhline(0)
ax1.axvline(pbar_uga.loc[my_j])
ax1.set_title(f"Uganda's Compensating Variation vs Revenue from Price Change for {my_j}")
```

```python
fig, ax1 = plt.subplots()

ax1.plot(P,[deadweight_loss(U0,pbar_uga,my_prices(p0,j=my_j)) for p0 in P])
ax1.set_xlabel("Price of %s" % my_j)
ax1.set_ylabel("Deadweight Loss")
ax1.set_title(f"Uganda's Deadweight Loss between revenue and compensating variation for {my_j}")
```

## The Specific Micronutrients of Focus: Vitamin A, Vitamin C, Vitamins B-9, and B-1

Micronutrients refers to vitamins and minerals, which are essential for the function of the body, fighting disease, and overall healthy development of the body.

## 7. <span style="color:red">*Deliverable 1.8[B]*</span> Policy Options <a name="policy-options"></a>

![GMOs](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC10171133/bin/KGMC_A_2208999_F0001_OC.jpg)

In this section, we will explore a variety of different GMO-related policies. Based upon research, we found that within these countries, GMO use is illegal. Our research aims to fight against this law and show how using GMOs within policy can greatly aid individuals of the population in meeting nutritional goals.

***In terms of policy, this would imply a mandate in utilizing GMO-based crops and foods within the society.***

In order to achieve everything, we utilize a function for Tanzania able to abstract the simulation, `simulate_gmo_tanzania`.

```python
def simulate_gmo_tanzania(crops: list, amount: float, nutrient: str, test_case = False):
    """
    TANZANIA
    Simulates the impact of GMO fortification on the intake of a specified nutrient
    by increasing the nutrient content in specified crops and visualizing the change
    in nutrient intake distribution before and after GMO fortification.

    Parameters:
    - crops (list of str): List of crop names in the FCT whose nutrient content is to be increased.
    - amount (float): The amount of the nutrient to add to each crop unit.
    - nutrient (str): The name of the nutrient to be fortified (e.g., 'Vitamin C').

    Outputs:
    - A histogram comparison of nutrient ratios before and after GMO fortification.
    - Prints the percentage of the population meeting or exceeding the recommended daily intake (RDI)
      for the nutrient, before and after fortification.
    """
    
    x_tanzania_gmo = pd.read_csv("data/tanzania/tanzania_expenditures_2019-2020.csv")
    if 'm' not in x_tanzania_gmo.columns:
        x_tanzania_gmo['m'] = 1
    x_tanzania_gmo = x_tanzania_gmo.set_index(['i', 't', 'm'])
    x_tanzania_gmo.columns.name = 'j'
    x_tanzania_gmo = x_tanzania_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    x_tanzania_gmo = x_tanzania_gmo.replace(0, np.nan)
    z_tanzania_gmo = pd.read_csv("data/tanzania/tanzania_HH_characteristics.csv")
    if 'm' not in z_tanzania_gmo.columns:
        z_tanzania_gmo['m'] = 1
    z_tanzania_gmo = z_tanzania_gmo.set_index(['i', 't', 'm'])
    z_tanzania_gmo.columns.name = 'k'
    z_tanzania_gmo = z_tanzania_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    p_tanzania_gmo = pd.read_csv("data/new.csv")
    if 'm' not in p_tanzania_gmo.columns:
        p_tanzania_gmo['m'] = 1
    p_tanzania_gmo = p_tanzania_gmo.set_index(['t', 'm'])
    p_tanzania_gmo.columns.name = 'j'
    p_tanzania_gmo = p_tanzania_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    p_tanzania_gmo = p_tanzania_gmo.replace(0, np.nan)
    fct_tanzania_gmo = pd.read_csv("data/tanzania/tanzania_fct.csv")
    fct_tanzania_gmo = fct_tanzania_gmo.set_index('j')
    fct_tanzania_gmo.columns.name = 'n'
    fct_tanzania_gmo = fct_tanzania_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))


    additional = amount
    for crop in crops:
        fct_tanzania_gmo.loc[crop, nutrient] += additional

    rdi_tanzania_gmo = pd.read_csv("data/uganda/uganda_rdi.csv")
    rdi_tanzania_gmo = rdi_tanzania_gmo.set_index('n')
    rdi_tanzania_gmo.columns.name = 'k'
    result_tanzania_gmo = cfe.Regression(y=np.log(x_tanzania_gmo.stack()), d=z_tanzania_gmo)
    xhat = result_tanzania_gmo .predicted_expenditures()
    qhat = (xhat.unstack('j')/p_tanzania_gmo).dropna(how='all')
    qhat = qhat.loc[:,qhat.count()>0]
    use = fct_tanzania_gmo.index.intersection(qhat.columns)
    nutrients = qhat[use]@fct_tanzania_gmo.loc[use,:]
    nutrients.mean() 

    z = z_tanzania_gmo[rdi_tanzania_gmo.columns.tolist()]

    hh_rdi = z@rdi_tanzania_gmo.T

    hh_rwi = hh_rdi*7

    use_nutrients = nutrients.columns.intersection(hh_rwi.columns)

    nutrient_ratio_tanzania_gmo = (nutrients[use_nutrients]/hh_rwi[use_nutrients]).dropna()
    
    plt.figure(figsize=(10, 6))

    # Plotting the normalized histogram for Uganda's energy nutrient ratio
    plt.hist(nutrient_ratio_tanzania[nutrient], bins=30, alpha=0.5, label='Tanzania', density=True, color="red")

    # Plotting the normalized histogram for Tanzania's energy nutrient ratio
    plt.hist(nutrient_ratio_tanzania_gmo[nutrient], bins=30, alpha=0.5, label='Tanzania with GMO', density=True, color="green")

    plt.title('Normalized Comparison of ' + nutrient + " Nutrient Ratios: Uganda vs Tanzania")
    plt.xlabel(nutrient + ' Nutrient Ratio')
    plt.ylabel('Density')  # Adjusting label to 'Density' for clarity
    plt.legend(loc='upper right')

    if not test_case:
        plt.show()

    percentage_meeting_rdi_before = (nutrient_ratio_tanzania[nutrient] >= 1).mean() * 100
    percentage_meeting_rdi_after = (nutrient_ratio_tanzania_gmo[nutrient] >= 1).mean() * 100

    print("Percentage of population meeting/exceeding RDI for " + nutrient + f" before GMO: {percentage_meeting_rdi_before:.2f}%")
    print("Percentage of population meeting/exceeding RDI for " + nutrient + f" after GMO: {percentage_meeting_rdi_after:.2f}%")

    if test_case:
        plt.close()
        return {
            'fct_tanzania_gmo': fct_tanzania_gmo,
            'nutrient_ratio_before': nutrient_ratio_tanzania[nutrient],
            'nutrient_ratio_after': nutrient_ratio_tanzania_gmo[nutrient],
            'percentage_meeting_rdi_before': percentage_meeting_rdi_before,
            'percentage_meeting_rdi_after': percentage_meeting_rdi_after
        }
```

```python
def simulate_gmo_uganda(crops: list, amount: float, nutrient: str, test_case = False):
    """
    UGANDA
    Simulates the impact of GMO fortification on the intake of a specified nutrient
    by increasing the nutrient content in specified crops and visualizing the change
    in nutrient intake distribution before and after GMO fortification.

    Parameters:
    - crops (list of str): List of crop names in the FCT whose nutrient content is to be increased.
    - amount (float): The amount of the nutrient to add to each crop unit.
    - nutrient (str): The name of the nutrient to be fortified (e.g., 'Vitamin C').

    Outputs:
    - A histogram comparison of nutrient ratios before and after GMO fortification.
    - Prints the percentage of the population meeting or exceeding the recommended daily intake (RDI)
      for the nutrient, before and after fortification.
    """
    
    x_uganda_gmo = pd.read_csv("data/uganda/uganda_expenditures_2019-2020.csv")
    if 'm' not in x_uganda_gmo.columns:
        x_uganda_gmo['m'] = 1
    x_uganda_gmo = x_uganda_gmo.set_index(['i', 't', 'm'])
    x_uganda_gmo.columns.name = 'j'
    x_uganda_gmo = x_uganda_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    x_uganda_gmo = x_uganda_gmo.replace(0, np.nan)
    z_uganda_gmo = pd.read_csv("data/uganda/uganda_HH_characteristics.csv")
    if 'm' not in z_uganda_gmo.columns:
        z_uganda_gmo['m'] = 1
    z_uganda_gmo = z_uganda_gmo.set_index(['i', 't', 'm'])
    z_uganda_gmo.columns.name = 'k'
    z_uganda_gmo = z_uganda_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    p_uganda_gmo = pd.read_csv("data/uganda/uganda_food_prices_2019-2020.csv")
    if 'm' not in p_uganda_gmo.columns:
        p_uganda_gmo['m'] = 1
    p_uganda_gmo = p_uganda_gmo.set_index(['t', 'm'])
    p_uganda_gmo.columns.name = 'j'
    p_uganda_gmo = p_uganda_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))
    p_uganda_gmo = p_uganda_gmo.replace(0, np.nan)
    fct_uganda_gmo = pd.read_csv("data/uganda/uganda_fct.csv")
    fct_uganda_gmo = fct_uganda_gmo.set_index('j')
    fct_uganda_gmo.columns.name = 'n'
    fct_uganda_gmo = fct_uganda_gmo.apply(lambda x: pd.to_numeric(x, errors='coerce'))

    additional = amount
    for crop in crops:
        fct_uganda_gmo.loc[crop, nutrient] += additional

    rdi_uganda_gmo = pd.read_csv("data/uganda/uganda_rdi.csv")
    rdi_uganda_gmo = rdi_uganda_gmo.set_index('n')
    rdi_uganda_gmo.columns.name = 'k'
    result_uganda_gmo = cfe.Regression(y=np.log(x_uganda_gmo.stack()), d=z_uganda_gmo)
    xhat = result_uganda_gmo.predicted_expenditures()
    qhat = (xhat.unstack('j')/p_uganda_gmo).dropna(how='all')
    qhat = qhat.loc[:,qhat.count()>0]
    use = fct_uganda_gmo.index.intersection(qhat.columns)
    nutrients = qhat[use]@fct_uganda_gmo.loc[use,:]
    nutrients.mean() 
    z = z_uganda_gmo[rdi_uganda_gmo.columns.tolist()]
    hh_rdi = z@rdi_uganda_gmo.T
    hh_rwi = hh_rdi*7
    use_nutrients = nutrients.columns.intersection(hh_rwi.columns)
    aligned_nutrients, aligned_hh_rwi = nutrients[use_nutrients].align(hh_rwi[use_nutrients], axis=0)
    nutrient_ratio_uganda_gmo = (aligned_nutrients / aligned_hh_rwi).dropna()

    plt.figure(figsize=(10, 6))
    plt.hist(nutrient_ratio_uganda[nutrient], bins=30, alpha=0.5, label='Uganda', density=True, color="red")
    plt.hist(nutrient_ratio_uganda_gmo[nutrient], bins=30, alpha=0.5, label='Uganda with GMO', density=True, color="green")
    plt.title('Normalized Comparison of ' + nutrient + ' Nutrient Ratios: Uganda vs Uganda w GMO')
    plt.xlabel(nutrient + ' Nutrient Ratio')
    plt.ylabel('Density')
    plt.legend(loc='upper right')

    if not test_case:
        plt.show()

    percentage_meeting_rdi_before = (nutrient_ratio_uganda[nutrient] >= 1).mean() * 100
    percentage_meeting_rdi_after = (nutrient_ratio_uganda_gmo[nutrient] >= 1).mean() * 100

    print("Percentage of population meeting/exceeding RDI for " + nutrient + f" before GMO: {percentage_meeting_rdi_before:.2f}%")
    print("Percentage of population meeting/exceeding RDI for " + nutrient + f" after GMO: {percentage_meeting_rdi_after:.2f}%")

    if test_case:
        plt.close()
        return {
            'fct_uganda_gmo': fct_uganda_gmo,
            'nutrient_ratio_before': nutrient_ratio_uganda[nutrient],
            'nutrient_ratio_after': nutrient_ratio_uganda_gmo[nutrient],
            'percentage_meeting_rdi_before': percentage_meeting_rdi_before,
            'percentage_meeting_rdi_after': percentage_meeting_rdi_after
        }
```
